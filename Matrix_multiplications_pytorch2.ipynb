# **2. Matrix Multiplication in PyTorch**


Matrix multiplication is essential for neural networks! PyTorch provides efficient operations for multiplying matrices and tensors.

```python
import torch
```

# **1.Matrix Multiplication**

```python
# We use @ operator for matrix multiplication and we use torch.matmul


print("Matrix Multiplication By @operator ...")
A = torch.tensor([[1,2],[3,4]])

B = torch.tensor([[5,6],[7,8]])

print(A @ B)

print("Matrix Multiplication By MatMul...")
mat_mult = torch.matmul(A,B)
print(mat_mult)
# print(type(mat_mult))
```

# **# 2. MatrixVector Multiplication**

```python
M = torch.tensor([[1,2],[3,4]],dtype = torch.float32)
print(M)
print(M.shape)

V = torch.tensor([5,6],dtype = torch.float32)

print(V)
print(V.shape)

result = M @ V
print("M @ V matrix-vector Multiplication :")
print("\n result :",result)
print(result.shape)
```

# **3. Batch Matrix Multiplication**

```python
batch_a = torch.randn(3,2,4)
batch_b = torch.randn(3,4,3)
print("The batch_a Matrix ")
print(batch_a)
print(batch_a.shape)

print("The batch_b matrix ")
print(batch_b)
print(batch_b.shape)


print("The multiplication b/w  A and B ")
batch_result = torch.bmm(batch_a,batch_b)
print(batch_result)

# A0 = batch_a[0]
# B0 = batch_b[0]

# result0 = A0 @ B
# print(result0)
```

# **4. Element-wise Multiplication**

```python
A = torch.tensor([[1,2],[3,4]],dtype = torch.float32)
print("\n The matrix A ",A)
B = torch.tensor([[5,6],[7,8]],dtype = torch.float32)
print("\n The matrix B ",B)
print("The Element wise matrix Multiplication B/W A and B ...")
print(A * B)
```
